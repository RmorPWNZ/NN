# Подготовим данные к обучению и обучим модель

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['MEDV'],axis=1), df['MEDV'], test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_,df.drop(['MEDV'],axis=1).columns,columns=['Coefficient'])
coeff_df
```

Интерпретировать можно как: при фиксированных всех значениях постоянными увеличение характеристики "RM"(среднее кол-во комнат) на 1 единицу увеличит цену дома на $3247.765. Аналогично для остальных параметров.
